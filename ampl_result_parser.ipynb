```python
import pathlib
import re

import pandas as pd

result_dir = pathlib.Path("ampl_outputs")
instance_dir = pathlib.Path("instances")
filepaths = list(instance_dir.glob("*"))
filepaths = [filepath for filepath in filepaths if ("ampl" in str(filepath) and not "taipei" in str(filepath))]
# print(filepaths)
```

```python
result_dict_list = []
for filepath in filepaths:
    filename = filepath.name
    result_filename = f"{filename}.out"
    result_filepath = result_dir/result_filename
    instance_name = result_filename[:-13]
    print(instance_name)
    with open(result_filepath.absolute(), "r") as f:
        text = f.read()
        tc = re.findall(r"TC:\s*(\d+(\.\d*)?)", text)[0][0]
        tc_bound = re.findall(r"TC\.best_bound:\s*(\d+(\.\d*))?", text)[0][0]
        runtime = re.findall(r"running_time:\s*(\d+(\.\d*)?)", text)[0][0]
        mip_gap =  float(re.findall(r"MIP Gap \(%\):\s*(\d+(\.\d*)?(e\-\d+)?)", text)[0][0])
        mrt_lines = re.findall(r"MRT used: (\d+ â†’ \d+)", text)
        num_mrt_lines_used = len(mrt_lines)
        result_dict = {
            "instance":instance_name,
            "TC":tc,
            "TC Bound":tc_bound,
            "Running time":runtime,
            "MIP Gap":mip_gap,
            "Num MRT Line Used":num_mrt_lines_used
        }
        # print(instance_name, mip_gap)
        result_dict_list.append(result_dict)
result = pd.DataFrame(result_dict_list)
```

```python
result.to_csv("ampl_result_summary.csv", index=False)
```
