```python
import pathlib
import os

results_dir = pathlib.Path("results_new_model")
result_filepaths = list(results_dir.glob("*"))
```

```python
result_dict_list = []
for result_filepath in result_filepaths:
    result_dict = {}
    text = ""
    instance_name = result_filepath.name[:-4]
    with open(result_filepath.absolute(), "r") as f:
        lines = f.readlines()
        params = lines[:6]
        for param in params:
            key, value = param.split(":")
            result_dict[key] = float(value)
        result_dict["INSTANCE NAME"] = instance_name
        result_dict_list.append(result_dict)
```

```python
import pandas as pd
df = pd.DataFrame(result_dict_list)
```

```python
df.to_csv("new_model_raw_results.csv")
```
